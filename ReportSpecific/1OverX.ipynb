```python
from geomdl import NURBS

# Create the curve instance
crv = NURBS.Curve()


# Set degree
crv.degree = 2

w = (121/40)**(0.5)
# Set control points
crv.ctrlptsw = [[1, 1, 0, 1], [w*20/11, w*18/99, 0, w], [10, 0.1, 0, 1]]

# Set knot vector
crv.knotvector = [0, 0, 0,  1, 1, 1]

# Import Matplotlib visualization module
from geomdl.visualization import VisMPL

# Set the visualization component of the curve
crv.vis = VisMPL.VisCurve2D()

# Plot the curve
crv.render()
```

```python
for i in range(100):
    m = crv.evalpts[i][0]*crv.evalpts[i][1] - 1
    print(m)
```
