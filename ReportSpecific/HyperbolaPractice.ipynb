```python
from geomdl import NURBS
import numpy as np
from geomdl.visualization import VisMPL
```

```python
T = 1
a = 2
b = 1
c = np.cosh(T)
s = np.sinh(T)
t = np.tanh(T)
w = np.cosh(T/2)
```

```python
crv = NURBS.Curve()
crv.degree = 2

crv.ctrlptsw = [[a, 0, 0, 1], [a*w, w*b*np.tanh(T/2), 0, w], [a*c, b*s, 0, 1]]
crv.knotvector = [0, 0, 0,  1, 1, 1]

crv.vis = VisMPL.VisCurve2D()
crv.render()
```

```python
for i in range(100):
    m = (crv.evalpts[i][0]/a)**2 - (crv.evalpts[i][1]/b)**2 -1
    print(m)
```
