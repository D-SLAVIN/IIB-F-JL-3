```python
from geomdl import NURBS
import numpy as np
from geomdl.visualization import VisMPL
from geomdl import multi
```

```python
T = 1
a = 1.944
b = 0.5106
c = np.cosh(T)
s = np.sinh(T)
t = np.tanh(T)
w3 = c
w5 = np.cosh(T/2)
```

```python
t = 10
w = 1
A = 0.03
```

```python
crv3h = NURBS.Curve()
crv3h.degree = 2

crv3h.ctrlptsw = [[a*c, -b*s, 0, 1], [w3*a/c, 0, 0, w3], [a*c, b*s, 0, 1]]
crv3h.knotvector = [0, 0, 0,  1, 1, 1]

crv3p = NURBS.Curve()
crv3p.degree = 2

crv3p.ctrlptsw = [[A*t**2, -2*A*t, 0, 1], [-A*t**2, 0, 0, w], [A*t**2, 2*A*t, 0, 1]]
crv3p.knotvector = [0, 0, 0,  1, 1, 1]

mcrv = multi.CurveContainer(crv3h, crv3p)
mcrv.vis = VisMPL.VisCurve2D(figure_size=[6,4], axes_equal=False, figure_dpi=144)
mcrv.render(evalcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], cpcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], filename="HyperbolaComp.pdf", plot=True )
```
