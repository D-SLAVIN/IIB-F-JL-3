```python
from clifford.tools.g3c import *
from clifford.tools.g3c.rotor_parameterisation import *
from pyganja import *
import numpy as np
```

```python
BR = e1^e2
BS = e4^e5
N = 10 
f= 0.25
a = 2
b = 1
R = general_exp(-BR*f*np.pi/N)
X0 = up(e1)

circle_arc = [R**(i)*X0*~R**(i) for i in range(N+1)]



e = (1-b**2/a**2)**0.5
r = [b/(1-(e*np.cos(i*2*np.pi*f/N))**2)**(0.5) for i in range(N+1)]

elliptical_arc = [general_exp(-BS*np.log(r[i])/2)*circle_arc[i]*~(general_exp(-BS*np.log(r[i])/2)) for i in range(N+1)]
```

```python
BRev = e1^e3
M = 10 
frev= 0.5
Rrev = general_exp(-BRev*frev*np.pi/M)

surf = [[Rrev**(j)*elliptical_arc[i]*~Rrev**(j) for j in range(M+1)] for i in range(N+1)]
pp_list = [[surf[i][j]^surf[i][j+1] for i in range(N+1)] for j in range(M)]
pp_list2 = [[surf[i][j]^surf[i+1][j] for j in range(M+1)] for i in range(N)]
```

```python
sc = GanjaScene()

sc.add_objects([normalise_n_minus_1(o(1)).clean(1E-6) for o in elliptical_arc], color=Color.BLACK)
for i in range(N):
    sc.add_objects([o(2).normal().clean(1E-4) for o in pp_list[i]], color=Color.BLACK)
for j in range(M):
    sc.add_objects([o(2).normal().clean(1E-4) for o in pp_list2[j]], color=Color.BLACK)
draw(sc, 
    static=False, 
    scale=0.1, new_window=False)
```
