```python
from geomdl import NURBS
import numpy as np
from geomdl.visualization import VisMPL
from geomdl import multi
```

```python
t = 10
w = 1
A = 0.03
```

```python
crv3 = NURBS.Curve()
crv3.degree = 2

crv3.ctrlptsw = [[A*t**2, -2*A*t, 0, 1], [-A*t**2, 0, 0, w], [A*t**2, 2*A*t, 0, 1]]
crv3.knotvector = [0, 0, 0,  1, 1, 1]

```

```python
crv5 = NURBS.Curve()
crv5.degree = 2

crv5.ctrlptsw = [[A*t**2, -2*A*t, 0, 1],  [0, -A*t, 0, w],
                 [0, 0, 0, w],
                 [0, A*t, 0, w], [A*t**2, 2*A*t, 0, 1]]
crv5.knotvector = [0, 0, 0, 0.5, 0.5,  1, 1, 1]

```

```python
crv9 = NURBS.Curve()
crv9.degree = 2

crv9.ctrlptsw = [[A*t**2, -2*A*t, 0, 1], [A*t**2/2, -3*A*t/2, 0, 1], [A*t**2/4, -A*t, 0, 1], [0, -A*t/2, 0, w],
                 [0, 0, 0, w],
                 [0, A*t/2, 0, w], [A*t**2/4, A*t, 0, 1], [A*t**2/2, 3*A*t/2, 0, 1], [A*t**2, 2*A*t, 0, 1]]
crv9.knotvector = [0, 0, 0, 0.25, 0.25, 0.5, 0.5, 0.75, 0.75,  1, 1, 1]

crv3.name = 'n=3'
crv5.name = 'n=5'
```

```python
mcrv = multi.CurveContainer(crv3, crv5)
mcrv.vis = VisMPL.VisCurve2D(figure_size=[6,4], axes_equal=False, figure_dpi=144)
mcrv.render(evalcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], cpcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], filename="ParabolaComp.pdf", plot=True )
```
