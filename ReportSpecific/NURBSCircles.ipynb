```python
from geomdl import NURBS
from geomdl.visualization import VisMPL
import numpy as np
from geomdl import multi
import matplotlib.pyplot as plt
```

```python
theta = 4*np.pi/6
w1 = np.cos(theta/2)
crv1 = NURBS.Curve()
crv1.degree = 2
crv1.ctrlptsw = [[1, 0, 0, 1], [1*w1, np.tan(theta/2)*w1, 0, w1], [np.cos(theta), np.sin(theta), 0, 1]]
crv1.knotvector = [0, 0, 0,  1, 1, 1]

crv1.name = 'Curve 1'

w2 = np.cos(theta/4)
c = np.cos(theta/4)
s = np.sin(theta/4)
t = np.tan(theta/4)
crv2 = NURBS.Curve()
crv2.degree = 2
crv2.ctrlptsw = [[1, 0, 0, 1], [1*w2, np.tan(theta/4)*w2, 0, w2], [np.cos(theta/2), np.sin(theta/2), 0, 1], [(c**2 - 3*s**2)*w2, (3*c*s-s**2*t)*w2, 0, w2], [np.cos(theta), np.sin(theta), 0, 1]]
crv2.knotvector = [0, 0, 0, 0.5, 0.5,  1, 1, 1]

crv2.name = 'Curve 2'

mcrv = multi.CurveContainer(crv1, crv2)
mcrv.vis = VisMPL.VisCurve2D(figure_size=[6,4], axes_equal=True, figure_dpi=144)
mcrv.render(evalcolor= [(0, 0, 0, 1),(0, 0, 1, 1)], cpcolor= [(0, 0, 0, 1),(0, 0, 1, 1)], filename="Arc5pi6.pdf", plot=True )
```

```python
crv3 = NURBS.Curve()
crv3.degree = 2
crv3.ctrlptsw = [[0, 1, 0, 1], [0, 0.5*2, 0, 0.5], [-0.5*2, 0.5*2, 0, 0.5], [-2, 1, 0, 1],
                [-0.5*2, 0, 0, 0.5], [0, 0, 0, 0.5], [0, 1, 0, 1]]
crv3.knotvector = [0, 0, 0, 0.25, 0.5, 0.5, 0.75, 1, 1, 1]

crv3.name = 'Curve 1'

crv4 = NURBS.Curve()
crv4.degree = 2
w = (1/2)**(0.5)
crv4.ctrlptsw = [[+2, -1, 0, 1], [+2*w, 0, 0, w], [+1, 0, 0, 1], [0, 0, 0, w], [0, -1, 0, 1],
                [0, -2*w, 0, w], [+1, -2, 0 , 1], [+2*w, -2*w, 0, w], [+2, -1, 0, 1]]
crv4.knotvector = [0, 0, 0, 0.25, 0.25, 0.5, 0.5, 0.75, 0.75, 1, 1, 1]

crv4.name = 'Curve 2'

mcrv2 = multi.CurveContainer(crv3, crv4)
mcrv2.vis = VisMPL.VisCurve2D(figure_size=[6,4], axes_equal=True, figure_dpi=144)
mcrv2.render(evalcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], cpcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], filename="FullCircleComp.pdf", plot=True )
```
