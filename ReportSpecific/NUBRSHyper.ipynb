```python
from geomdl import NURBS
import numpy as np
from geomdl.visualization import VisMPL
from geomdl import multi
```

```python
T = 1
a = 1.944
b = 0.5106
c = np.cosh(T)
s = np.sinh(T)
t = np.tanh(T)
w3 = c
w5 = np.cosh(T/2)
```

```python
crv3 = NURBS.Curve()
crv3.degree = 2

crv3.ctrlptsw = [[a*c, -b*s, 0, 1], [w3*a/c, 0, 0, w3], [a*c, b*s, 0, 1]]
crv3.knotvector = [0, 0, 0,  1, 1, 1]

crv5 = NURBS.Curve()
crv5.degree = 2

crv5.ctrlptsw = [[a*c, -b*s, 0, 1], [a*w5, -w5*b*np.tanh(T/2), 0, w5], 
                 [a, 0, 0, 1], [a*w5, w5*b*np.tanh(T/2), 0, w5], [a*c, b*s, 0, 1]]
crv5.knotvector = [0, 0, 0, 0.5, 0.5,  1, 1, 1]

crv3.name = 'n=3'
crv5.name = 'n=5'

mcrv = multi.CurveContainer(crv3, crv5)
mcrv.vis = VisMPL.VisCurve2D(figure_size=[6,4], axes_equal=True, figure_dpi=144)
mcrv.render(evalcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], cpcolor= [(0, 0, 0, 1),(0, 0, 0.7, 1)], filename="NURBSHyperbola.pdf", plot=True )
```
