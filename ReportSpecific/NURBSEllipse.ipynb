```python
from geomdl import NURBS
import numpy as np
from geomdl.visualization import VisMPL
from geomdl import multi
```

```python
w = (1/2)**(0.5)
crv1 = NURBS.Curve()
crv1.degree = 2
crv1.ctrlptsw = [[1, 0, 0, 1], [1*w, 1*w, 0, w], [0, 1, 0, 1]]
crv1.knotvector = [0, 0, 0,  1, 1, 1]
crv1.name = 'Circular Arc'


crv2 = NURBS.Curve()
crv2.degree = 2
crv2.ctrlptsw = [[2, 0, 0, 1], [2*w, 1*w, 0, w], [0, 1, 0, 1]]
crv2.knotvector = [0, 0, 0,  1, 1, 1]
crv2.name = 'Elliptical Arc'


mcrv = multi.CurveContainer(crv1, crv2)
mcrv.vis = VisMPL.VisCurve2D(figure_size=[6,4], axes_equal=True, figure_dpi=144)
mcrv.render(evalcolor= [(0, 0, 0, 1),(0, 0, 1, 1)], cpcolor= [(0, 0, 0, 1),(0, 0, 1, 1)], filename="NURBSEllipse.pdf", plot=True )
```
