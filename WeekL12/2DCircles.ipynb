# Quarter Circle 
Degree 2 from https://www.geometrictools.com/Documentation/NURBSCircleSphere.pdf

```python
from geomdl import NURBS

# Create the curve instance
crv = NURBS.Curve()

# Set degree
crv.degree = 2

# Set control points
crv.ctrlptsw = [[1, 0, 0, 1], [0.707, 0.707, 0, 0.707], [0, 1, 0, 1]]

# Set knot vector
crv.knotvector = [0, 0, 0,  1, 1, 1]

# Import Matplotlib visualization module
from geomdl.visualization import VisMPL

# Set the visualization component of the curve
crv.vis = VisMPL.VisCurve3D()

# Plot the curve
crv.render()
```

# Full Circle 
Degree 2 from https://pages.mtu.edu/~shene/COURSES/cs3621/NOTES/spline/NURBS/RB-circles.html

```python
from geomdl import NURBS

# Create the curve instance
crv = NURBS.Curve()

# Set degree
crv.degree = 2

# Set control points
crv.ctrlptsw = [[1, 0, 0, 1], [0.707, 0.707, 0, 0.707], [0, 1, 0, 1], [-0.707, 0.707, 0, 0.707], [-1, 0, 0, 1], [-0.707, -0.707, 0, 0.707], [0, -1, 0 , 1], [0.707, -0.707, 0, 0.707], [1, 0, 0, 1]]

# Set knot vector
crv.knotvector = [0, 0, 0, 0.25, 0.25, 0.5, 0.5, 0.75, 0.75, 1, 1, 1]

# Import Matplotlib visualization module
from geomdl.visualization import VisMPL

# Set the visualization component of the curve
crv.vis = VisMPL.VisCurve3D()

# Plot the curve
crv.render()
```
