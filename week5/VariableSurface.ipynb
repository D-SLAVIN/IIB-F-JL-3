```python
from clifford.tools.g3c import *
from clifford.tools.g3c.rotor_parameterisation import *
from pyganja import *
import numpy as np
```

```python
#set up point pairs from (5,0,0) to (15,0,0) and one from (0,15,0) to (0,25,0)
#the same as a anticlockwise rotation of pi/4 in the e1^e2 plane and a translation of 10 along the e2 axis

x1 = 5*e1
X1 = up(x1)
y1 = 15*e1
Y1 = up(y1)

line1 = X1^Y1

x2 = 15*e2
X2 = up(x2)
y2 = 25*e2
Y2 = up(y2)

line2 = X2^Y2





#Construct rotor and lower by N/2

N = 40
B = e1^e2
R = general_exp(-B*math.pi/(4*(N/2)))


#form an array from 0 to N-1
#form an array of equal length by random
                
order = np.arange(N)
rand = np.arange(N)
np.random.shuffle(rand)

#form an empty list of inplane lines

plane_list = [0]*(N+1)
plane_list[0] = line1

#Build up the list
for i in order:
    #If random number is even apply a rotation
    if rand[i] % 2 == 0 :
        plane_list[i+1] = R*plane_list[i]*~R
        
    #else apply a transformation 
    else :
        F = plane_list[i].normal()
        P = 0.5*(1 + F)
        A = P*((plane_list[i])|einf)
        a = down(A)
        B = - ~P*((plane_list[i])|einf)
        b = down(B)
        T = general_exp(10*einf*a.normal()/(2*(N/2)))
        
        plane_list[i+1] = T*plane_list[i]*~T
        


        
#Create outside and inside point list
F = [plane_list[i].normal() for i in range(N+1)]

P = [0.5*(1 + F[i]) for i in range(N+1)]

OPoint_list = [ P[i]*((plane_list[i])|einf) for i in range(N+1)]

IPoint_list = [-(~P[i])*((plane_list[i])|einf) for i in range(N+1)]

#Create Edge Lists

OEdge_list = [OPoint_list[i]^OPoint_list[i+1] for i in range(N)]
IEdge_list = [IPoint_list[i]^IPoint_list[i+1] for i in range(N)]


sc = GanjaScene()

sc.add_objects([o(2).normal().clean(1E-6) for o in plane_list], color=Color.BLUE)
sc.add_objects([o(2).normal().clean(1E-6) for o in OEdge_list], color=Color.BLUE)
sc.add_objects([o(2).normal().clean(1E-6) for o in IEdge_list], color=Color.BLUE)

draw(sc, 
    static=False , 
    scale=0.1)
```
