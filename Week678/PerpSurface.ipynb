```python
from clifford.tools.g3c import *
from clifford.tools.g3c.rotor_parameterisation import *
from pyganja import *
```

```python
#Set up a point pair 

z1 = 5*e3
Z1 = up(z1)

z2 = -5*e3
Z2 = up(z2)

L = Z2^Z1


#Parameters for the shape

N = 100
Len = 30
Rot = 1

#Set up rotors
B = e3*e2

R = general_exp(-Rot*B*math.pi/(N/2))
T = general_exp(Len*einf*e1/(2*(N/2)))



order = np.arange(N)
rand = np.arange(N)
#np.random.shuffle(rand)

line_list = [0]*(N+1)
line_list[0] = L



#Build up the list
for i in order:
    #If random number is even apply a rotation
    if rand[i] % 2 == 0 :
        line_list[i+1] = R*line_list[i]*~R
    else :
        line_list[i+1] = T*line_list[i]*~T
        
#Create outside and inside point list
F = [line_list[i].normal() for i in range(N+1)]

P = [0.5*(1 + F[i]) for i in range(N+1)]

OPoint_list = [ P[i]*((line_list[i])|einf) for i in range(N+1)]

IPoint_list = [-(~P[i])*((line_list[i])|einf) for i in range(N+1)]

#Create Edge Lists

OEdge_list = [OPoint_list[i]^OPoint_list[i+1] for i in range(N)]
IEdge_list = [IPoint_list[i]^IPoint_list[i+1] for i in range(N)]
        
        
        
sc = GanjaScene()

sc.add_objects([o(2).normal().clean(1E-6) for o in line_list], color=Color.BLUE)
sc.add_objects([o(2).normal().clean(1E-6) for o in OEdge_list], color=Color.BLUE)
sc.add_objects([o(2).normal().clean(1E-6) for o in IEdge_list], color=Color.BLUE)


draw(sc, 
    static=False , 
    scale=0.1)
```
